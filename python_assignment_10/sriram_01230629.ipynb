```python
import pandas as pd

data = pd.read_csv('./EmployeeAttrition.csv')
```

```python
print(data.head())
```

```python
#Check the head of the DataFrame.
print(data.head())
```

```python
#Q1: Find the number of entries/rows and columns in the data.
# Ans:(1470, 35)
print(data.shape)
print("-"*100)

#Q2: What is the average Monthly Income?
# Ans: 6502.931292517007
print(data.MonthlyIncome.mean())
print("-"*100)

#Q3: What is the highest amount of HourlyRate ?
# Ans:100
print(data.HourlyRate.max())
print("-"*100)

#Q4: What is the Department, JobRole, MaritalStatus and OverTime of EmployeeNumber 10?
# Ans:               Department                JobRole MaritalStatus
#     6  Research & Development  Laboratory Technician       Married
print(data[data['EmployeeNumber']==10][['Department','JobRole','MaritalStatus']])
print("-"*100)

#Q5: What is the Employee ID of highest MonthlyIncome paid employee?
# Ans:190    259
#   Name: EmployeeNumber, dtype: int64

print(data.sort_values(by=["MonthlyIncome"], ascending=False)["EmployeeNumber"][:1])
print("-"*100)

#Q6: What is the average(mean) DailyRate group by Age for all Employees whose age is greater than 58. (hint: use groupby function)
# Ans:Age
#     59    992.3
#     60    833.2
#Name: DailyRate, dtype: float64
df_age_gt_58 = data.loc[data['Age'] > 58]
print(df_age_gt_58.groupby(["Age"])["DailyRate"].mean())
print("-"*100)

#Q7: How many unique EducationField are there?
# Ans:['Life Sciences' 'Other' 'Medical' 'Marketing' 'Technical Degree'
# 'Human Resources']
print(data.EducationField.unique())
print("-"*100)

#Q8: What are the top 5 most common JobRole?
# Ans:Sales Executive              326
#Research Scientist           292
#Laboratory Technician        259
#Manufacturing Director       145
#Healthcare Representative    131
#Name: JobRole, dtype: int64
print(data['JobRole'].value_counts().head(5))
print("-"*100)

#Q9: How many JobRoles represented by less than 100 employees?
# Ans:JobRole
#Human Resources         52
#Research Director       80
#Sales Representative    83
#Name: EmployeeNumber, dtype: int64
emp_count = data.groupby(["JobRole"])["EmployeeNumber"].count()
print(emp_count[emp_count < 100])
print("-"*100)

#Q10: What is the correlation between Education and JobSatisfaction?
# Ans: -0.011296116736573911
print(data["Education"].corr(data["JobSatisfaction"]))
print("-"*100)
```
